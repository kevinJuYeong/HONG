```python
from PIL import Image
import matplotlib.pyplot as plt

im = Image.open('./images/parrot.png')
im2 = Image.open('./images/apple.png')
im3 = Image.open('./images/birds.png')
im4 = Image.open('./images/butterfly.png')

plt.figure(figsize=(8, 6))

plt.subplot(2, 2, 1)
plt.imshow(im)

plt.subplot(2, 2, 2)
plt.imshow(im2)

plt.subplot(2, 2, 3)
plt.imshow(im3)

plt.subplot(2, 2, 4)
plt.imshow(im4)

plt.show()
```
